```python
import pandas as pd
import sqlite3 as sql3
import re

conn = sql3.connect('jugadores.db')

df = pd.read_sql_query('SELECT * FROM mbappe_2 ORDER BY temporada', conn)
df
```

```python
# if 'edad' not in df.columns:
df['edad'] = pd.Series([pd.NA] * len(df), dtype='Int64')  # Inicializa con pd.NA y tipo Int64

temporadas = []
temporadas_activas = []
age = 15
temporada_anterior = '-'

# Recopilar todas las temporadas
for index, row in df.iterrows():
    temporada = row['temporada']  # el dato en cada celda, se guarda en la var
    temporadas.append(temporada)

    if pd.isnull(temporada) or not re.findall(r"[0-9]$", temporada):
        df.at[index, 'edad'] = pd.NA  # Usar pd.NA en lugar de 0
    else:  # a la lista se le agregael index y la temp actual en la iteracion
        temporadas_activas.append((index, temporada))

# Inicio primer temporada con 15 + 1 years old
for index, temporada in temporadas_activas:
    if temporada[:4] != temporada_anterior[:4]:
        age += 1
        df.at[index, 'edad'] = age  # 16 al index en la celda actual de edad
        temporada_anterior = temporada  # Actualizo temp anterior
    # Mantiene la edad si la temporada es la misma que la anterior
    elif temporada_anterior[:4] == temporada[:4]:
        df.at[index, 'edad'] = age

print(temporadas_activas)

df.to_sql('mbappe_2', conn, if_exists='replace', index=False, dtype={
    'club': 'TEXT',
    'temporada': 'TEXT',
    'partidos_liga': 'INTEGER',
    'goles_liga': 'INTEGER',
    'asistencias_liga': 'INTEGER',
    'partidos_copa': 'INTEGER',
    'goles_copa': 'INTEGER',
    'asistencias_copa': 'INTEGER',
    'partidos_internacional': 'INTEGER',
    'goles_internacional': 'INTEGER',
    'asistencias_internacional': 'INTEGER',
    'partidos_total': 'INTEGER',
    'goles_total': 'INTEGER',
    'asistencias_total': 'INTEGER',
    'media_goleadora': 'REAL',
    'edad': 'INTEGER'
})


conn.commit()
conn.close()
```
