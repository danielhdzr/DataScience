```python
import urllib.request
from bs4 import BeautifulSoup as bs
import re
import pandas as pd
```

```python
# Load HTML code from a URL
page = urllib.request.urlopen('https://docs.python.org/3/library/random.html')
# Pull HTML data from page with beautifulsoup
soup = bs(page)

# Con inspect vi que las funciones estan dentro de las etiquetas '<dt/dt>'
# Asi que voy a buscar todas las 'description term' tags o 'dt'
names = soup.body.findAll('dt')

# re.findall(look for, in) 
# Ya que 'names' encapsula el contenido HTML en el tag 'dt', hare mi siguiente busqueda ahi
# Busco codigo que comience con 'random.' y le sigue una serie de letras representadas con '\w+' en regex
# El contenido de ese codigo sera guardado en mi variable
function_names = re.findall(r'id="random.\w+', str(names)) # lo convierto a str para que sea interpretado como texto

# (List comprehension expresion) 
# Slice the first 4 caracters in each item or object or member in our variable
function_names = [item[4:] for item in function_names]

# Find all function description in the HTML I'm scraping
# Look for Description Details tag 'dd/dd' in the page with the inspect tool
description = soup.body.findAll('dd')

print(function_names) 
description # IT also prints this way
```

```python
# Remove the tags <>
function_usage = []
# En mi loop cada tag es un item
# El atributo '.text' da acceso al contenido dentro del objeto indicado. Funciona en XML, HTML y JSON
# Reemplazo los saltos de linea '\n' con espacios vacios ' '
for item in description:
    item = item.text.replace('\n', ' ')
    function_usage.append(item)

print(function_names)    
function_usage
```

```python
# Check if the lenght is the same in both lists
print(len(function_names))    
len(function_usage)
```

```python
# Create a dataframe 
# Pandas syntaxis, not python syntaxis
data = pd.DataFrame({'Function name': function_names, 'function usage': function_usage})
# Muestra solo los primeros cinco rows
data.head()
```
