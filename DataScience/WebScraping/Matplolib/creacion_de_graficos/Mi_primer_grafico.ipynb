```python
import pandas as pd
import matplotlib.pyplot as plt
# Creo mis ejes. 
# f(x) suele ser un nombre estandar para una funcion que se ejecuta sobre x, y regresa los valores en value. 
# En este caso no se definio f(x) y solo se usa como nombre, con values introducidos de forma manual, lo mismo en g(x).
# En el eje Y tengo dos lineas graficas.
y = {'f(x)' : [2, 10, 6, 8], 'g(x)' : [4, 2, -2, 6]}
x = [1, 2, 3, 4]

# Creo tabla
graph = pd.DataFrame(y, x)
# Creo mi grafico con valores en la tabla. 
# Elegi tipo linea, cuadriculado, con titulos
my_graph = graph.plot(kind = 'line', grid = True, title = 'My graph', ylabel = 'My Y title', xlabel = 'My X title')
# Guardo mi grafico
plt.savefig('mi_primer_grafico.pdf')
# Muestro mi grafico
plt.show()
```
