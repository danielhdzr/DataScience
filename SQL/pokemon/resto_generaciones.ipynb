```python
import pandas as pd
import sqlite3
import requests
from bs4 import BeautifulSoup
from io import StringIO

url = 'https://pokemon.fandom.com/es/wiki/Lista_de_Pokémon'
web_site = requests.get(url)

soup = BeautifulSoup(web_site.text, 'html.parser')

html_str = str(soup)

html_io = StringIO(html_str)
# El primer parametro de pd.read_html es el file path/
# A final de cuentas termina leyendo la url
tables = pd.read_html(html_io)

conn = sqlite3.connect('pokemon.db')

for index, table in enumerate(tables):
    print('*****************************************************')
    print('*****************************************************')
    print(f'index {index}')
    print(f'table {table}')
```
